### After cleaning we can start looking at the actual data

```python
two_p_csv = []
p_csv = []
hz_csv = []
s_csv = []
other = []
folder = '.\\med_MW'
# Iterate through all files in the directory
for filename in os.listdir(folder):
    # Check the file extension
    if filename.endswith("2p.csv"):
        two_p_csv.append(filename)
    elif filename.endswith("p.csv"):
        p_csv.append(filename)
    elif filename.endswith("hz.csv"):
        hz_csv.append(filename)
    elif filename.endswith("s.csv"):
        s_csv.append(filename)
    else:
        other.append(filename)
# Removing the files which we cannot read properly 
two_p_csv.remove('tsf10ms2p.csv')
hz_csv = ["tsf1hz.csv"]
```

### Original files :

```python
otherfiles = make_object_list(folder, other, show_fit=True)
mulotherfiles = MultipleOriginalFiles(otherfiles)
```

```python
from copy import deepcopy
df = deepcopy(mulotherfiles.dataframe)
```

```python
from data_viz_original import filter_dataframe
df['better_names'] = df['name'].str.replace('tsf', '').str.replace('.csv', '')
df['split'] = df['better_names'].str.split(',')
df['Vgs'] = df['split'].apply(lambda x: x[0])
df['Vds'] = df['split'].apply(lambda x: x[1])
df = filter_dataframe(df, thres=0.5)
pivoted = df.pivot_table(index='Vds', columns='Vgs', values='t1')
interpolated = pivoted.interpolate(method='linear', axis=0)
interpolated.plot(ylabel='t1 (ms)')
```
